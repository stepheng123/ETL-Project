```python
import pandas as pd 
tsv_file='une_rt_m.tsv'
csv_table=pd.read_table(tsv_file,sep='\t')
csv_table
```

```python
csv_table['Country'] = csv_table['s_adj,age,unit,sex,geo\\time'].apply(lambda x: x.split(",")[-1])
csv_table['AgeGroup'] = csv_table['s_adj,age,unit,sex,geo\\time'].apply(lambda x: x.split(",")[1])
csv_table['Unit'] = csv_table['s_adj,age,unit,sex,geo\\time'].apply(lambda x: x.split(",")[2])
csv_table['Gender'] = csv_table['s_adj,age,unit,sex,geo\\time'].apply(lambda x: x.split(",")[3])
csv_table['Seasonal_Adjustment'] = csv_table['s_adj,age,unit,sex,geo\\time'].apply(lambda x: x.split(",")[0])
```
